# Занятие № 3. Выборки из конечных совокупностей
### 3.1 
 Сначала найдем распределение частот признака X:
\begin{array}{rr} \hline
x_i &1 &2 &3 \\ \hline
n_i &3 &6 &1\\ \hline
\end{array}
$N$ - объем генеральной совокупности, в нашем случае $N = 10$


Затем вычислим генеральное среднее $\overline x_0$ признака $X$:
\begin{equation}
\overline x_0 = \frac{\sum\limits_{i = 1}^Nx_{i}n_i}{N} = \frac{x_{1}n_1 + x_{2}n_2 + x_{3}n_3}{N} = \frac{1*3 + 2*6 + 3*1}{10} = \frac{3 + 12 + 3}{10} = \frac{18}{10} = 1,8
\end{equation}
И генеральную дисперсию $Var(X)$:
\begin{equation}
Var(X) = \frac{1}{N}\sum\limits_{i = 1}^N(x_{0i} - \overline x_0)^2 = \frac{1}{10}(1^2*3 + 2^2*6 + 3^2*1) -  (\overline x_0)^2 = \frac{3 + 4*6 + 9}{10} - (1,8)^2 = \frac{3+24+9}{10} - 3,24 = 3,6 - 3,24 = 0,36
\end{equation}
По условию задачи имеем бесповторную выборку объема $n = 5$.


Математическое ожидание среднего значения $\overline X$ признака $X$ в данной выборке определим по теореме ( верно и для повторной выборки):
\begin{equation}
E(\overline X) = \overline x_0 = 1,8
\end{equation}
А дисперсию  среднего значения $\overline X$ признака $X$, в свою очередь, как: 
\begin{equation}
Var(\overline X) = \frac{Var(X)}{n}\frac{N-n}{N-1} = \frac{0,36}{5}\frac{10-5}{10-1} = \frac{0,36}{5}\frac{5}{9} = 0,04
\end{equation}
Ответ: $E(\overline X)= 1,8$; $Var(\overline X) = 0,04$.

### 3.2
Сначала найдем распределение частот признака X:
\begin{array}{rr} \hline
x_i &1 &2 &3 \\ \hline
n_i &2 &2 &6\\ \hline
\end{array}
$N$ - объем генеральной совокупности, в нашем случае $N = 10$


Затем вычислим генеральное среднее $\overline x_0$ признака $X$:
\begin{equation}
\overline x_0 = \frac{\sum\limits_{i = 1}^Nx_{i}n_i}{N} = \frac{x_{1}n_1 + x_{2}n_2 + x_{3}n_3}{N} = \frac{1*2 + 2*2 + 3*6}{10} = \frac{2 + 4 + 18}{10} = \frac{24}{10} = 2,4
\end{equation}
И генеральную дисперсию $Var(X)$:
\begin{equation}
Var(X) = \frac{1}{N}\sum\limits_{i = 1}^N(x_{0i} - \overline x_0)^2 = \frac{1}{10}(1^2*2 + 2^2*2 + 3^2*6) -  (\overline x_0)^2 = \frac{2 + 8 + 54}{10} - (2,4)^2 = \frac{64}{10} - 5,76 = 6,4 - 5,76 = 0,64
\end{equation}
 По условию задачи имеем бесповторную выборку объема $n = 5$. 
 
 
Математическое ожидание среднего значения $\overline X$ признака $X$ в данной выборке определим по теореме ( верно и для бесповторной выборки):
\begin{equation}
E(\overline X) = \overline x_0 = 2,4
\end{equation}
А дисперсию  среднего значения $\overline X$ признака $X$, в свою очередь, как: 
\begin{equation}
Var(\overline X) = \frac{Var(X)}{n} = \frac{0,64}{5} = 0,128
\end{equation}
Ответ: $E(\overline X)= 2,4$; $Var(\overline X) = 0,128$.

### 3.4
Согласно условию задачи имеем генеральную совокупность $N = 36$, элементами которой являются элементарные исходы, представляющие собой количество очков на красной и синей игральной костях после их одновременного подбрасывания, и выборочную совокупность объемом $n = 19$. 

Данные совокупности имеют 2 общих признака $X_1$ (число очков на красной кости) и $X_2$ (число очков на синей кости), причем $S = X_1 + X_2$. 

Кроме того, распределение $X_1$ и $X_2$ имеют одинаковое распределение:
\begin{array}{rr} \hline
X_{i} &1 &2 &3 &4 &5 &6 \\ \hline
P(X_i = j) &1/6 &1/6 &1/6 &1/6 &1/6 &1/6\\ \hline
\end{array} где $i = 1,2$; $j = \overline{1,...,6}$

Тогда генеральное среднее и генеральная дисперсия признаков $X_1$ и $X_2$ определяются следующим образом:
\begin{equation}
E(X_1) = E(X_2) = \overline x_{01} = \overline x_{02} = \frac{1}{6}*(1 + 2 + 3 + 4 + 5 + 6) = \frac{21}{6} = 3,5
\end{equation}
\begin{equation}
Var(X_1) = Var(X_2) = E(X_1^2) - (E(X_1))^2 = E(X_2^2) - (E(X_2))^2 = \frac{1 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2}{6} - (3,5)^2 = \frac{35}{12}
\end{equation}
Отсюда получим значения соотвествующих генарльных характеристик для признака $S$:
\begin{equation}
E(S) = x_{0s} = E(X_1 + X_2) = E(X_1) + E(X_2) = 2*E(X_i) = 2*3,5 = 7
\end{equation}
\begin{equation}
Var(S) = Var(X_1 + X_2) = Var(X_1) + Var(X_2) = 2*Var(X_i) = 2*\frac{35}{12} = \frac{35}{6}
\end{equation}
При вычислении генаральной дисперсии для $S$  использовали свойство дисперсии суммы независимых случайных величин $X_1$ и $X_2$ (число очков на i-ой игральной кости при их подбрасывании).

Используя формулы задачи $3.1$ для бесповторной выборки объема $n$ найдем математическое ожидание $E(\overline S)$ и дисперсию $Var(\overline S)$ среднего значения $\overline S$:
\begin{equation}
E(\overline S) = E(S) = x_{0s} = 7
\end{equation}
\begin{equation}
Var(\overline S) = \frac{Var(S)}{n}\frac{N-n}{N-1} = \frac{35}{6*19}\frac{17}{35} = \frac{17}{114} \approx 0,1491
\end{equation}
Ответ: $E(\overline S) = 7$; $Var(\overline S) \approx 0,1491$.

### 3.5

В первую очередь определим объем генеральной совокупности, сложив число студентов из разных групп между собой:
$N = n1 + n2 + n3 = 20 + 18 + 19 = 57$ студентов - в 3 группах.

Далее найдем эмпирическое (общее) среднее для совокупности $n = 57$, разбитой на несколько группп:
\begin{equation}
\overline x = \frac{1}{n}\sum\limits_{i = 1}^k\overline x_in_i = \frac{1}{n}\sum\limits_{i = 1}^k \sum\limits_{j = 1}^{n_i}x_{ij} =>
\end{equation}
\begin{equation}
=>
\overline x = \frac{\overline x_1n_1 + \overline x_2n_2 + \overline x_3n_3}{n1 + n2 + n3} = \frac{20*79 + 18*64 + 19*79}{57} = \frac{1580 + 1152 + 1501}{57} = \frac{1411}{19} \approx 74,263
\end{equation}

Генеральную(общую) дисперсию $Var(X)$ можно представить в виде:
\begin{equation}
Var(X) = \sigma^2 = \overline\sigma^2 + \delta^2
\end{equation}
Сначала найдем среднюю групповую дисперсию $\overline\sigma^2$:
\begin{equation}
\overline\sigma^2 = \frac{1}{n}\sum\limits_{i = 1}^k\sigma_i^2n_i = \frac{\sigma_1^2n_1 + \sigma_2^2n_2 + \sigma_3^2n_3}{n_1 + n_2 + n_3} = \frac{6^2*20 + 14^2*18 + 19^2*19}{57} = \frac{11107}{57} \approx 194,86
\end{equation}

Затем межгрупповую дисперсию $\delta^2$:
\begin{equation}
\delta^2 = \frac{1}{n}\sum\limits_{i = 1}^k(\overline x_i - \overline x)^2n_i = \frac{n_1(\overline x_1 - \overline x)^2 + n_2(\overline x_2 - \overline x)^2 + n_3(\overline x_3 - \overline x)^2}{N} = \frac{20*(79 - 74,263)^2 + 18*(64 - 74,263)^2 + 19*(79 - 74,263)^2}{57} = \frac{17550}{361} \approx 48,615
\end{equation}

Тогда генеральная дисперсия $Var(X)$ равна:
\begin{equation}
Var(X) =  \overline\sigma^2 + \delta^2 \approx 194,86 + 48,615 = 243,475
\end{equation}

В нужной нам аудитории находилось 20 студентов, значит, объем бесповторной (в каждой аудитории - разные студенты, ни один не сидит в 2-х и более аудиториях одновременно) выборки будет $n = 20$. 

Для нее найдем математическое ожидание и дисперсию по признаку $\overline X$ - среднему баллу за экзамен:
\begin{equation}
E(\overline X) = \overline x \approx  74,263
\end{equation}
\begin{equation}
Var(\overline X) = \frac{Var(X)}{n}\frac{N-n}{N-1} = \frac{243,475}{20}\frac{57-20}{57-1} = \frac{243,475}{20}\frac{37}{56} \approx 8,043
\end{equation}
Ответ: $E(\overline X)\approx 74,263$; $Var(\overline X) \approx 8,043$.

### 3.6
Распределение частот признака $X$ имеет вид:
\begin{array}{rr} \hline
X_i &300 &600\\ \hline
n_i &54 &46\\ \hline
\end{array}
А признака $Y$:
\begin{array}{rr} \hline
Y_i &3 &5 &8\\ \hline
n_i &30 &33 &37\\ \hline
\end{array}
Объем генеральной совокупности для $X$ $N_1 = 54 + 46 = 100$; для $Y$ он составит $N_2 = 30 + 33 + 37 = 100$.


Генеральное среднее $\overline x_0$ для $X$:
\begin{equation}
\overline x_0 = \frac{1}{100}(300*54 + 600*46) = \frac{43800}{100} = 438 = E(X)
\end{equation}
Генеральное среднее $\overline y_0$ для $Y$:
\begin{equation}
\overline y_0 = \frac{1}{100}(3*30 + 5*33 + 8*37) = \frac{551}{100} = 5,51 = E(Y)
\end{equation}

Вычисляем ковариацию $Cov(X,Y)$ признаков $X$ и $Y$ в генеральной совокупности:
\begin{equation}
Cov(X,Y) = E(XY) - E(X)E(Y) = \frac{300}{100}(3*17 + 5*19 + 8*18) + \frac{600}{100}(3*13 + 5*14 + 8*19) - 5,51*438 = 3*290 + 6*261 - 2413,38 
= 870 + 1566 - 2413,38 = 22,62
\end{equation}
Наконец, находим ковариацию $Cov(\overline X,\overline Y)$ для бесповторной выборки объема $n = 8$:
\begin{equation}
Cov(\overline X,\overline Y) = \frac{Cov(X,Y)}{n}\frac{N-n}{N-1} = \frac{1}{8}*22,62*\frac{100-8}{100-1} = \frac{22,62}{8}*\frac{92}{99} = \frac{8671}{3300} \approx 2,628
\end{equation}
Ответ: $Cov(\overline X,\overline Y)\approx 2,628$.

\begin{equation}

Вычисляем ковариацию $Cov(X,Y)$ признаков $X$ и $Y$ в генеральной совокупности:

Cov(X,Y) = E(XY) - E(X)E(Y) = \frac{7}{2000}(2*100 + 4*400 + 6*200) + \frac{10}{2000}(2*300 + 4*100 + 6*900) - 8,95*4,7 = 7*1,5 + 10*3,2 - 42,065

Теперь находим ковариацию $Cov(\overline X,\overline Y)$ для повторной выборки объема $n = 800$:

Cov(\overline X,\overline Y) = \frac{Cov(X,Y)}{n} = \frac{0,435}{800}\approx 0,00054375
\end{equation}
Ответ: $Cov(\overline X,\overline Y)\approx 0,00054375$.

### 3.8
Введем случайные величины $X_1$ и $X_2$, показывающие поддерживает ли случайно взятый избиратель соответственно партию А или партию В.

Тогда распределение $X_1$ имеет вид:
\begin{array}{rr} \hline
X_1 &0 &1\\ \hline
P(A) &0,84 &0,16\\ \hline
\end{array}
А распределение $X_2$ имеет вид:
\begin{array}{rr} \hline
X_2 &0 &1\\ \hline
P(B) &0,77 &0,23\\ \hline
\end{array}

Вычислим ковариацию $Cov(X_1,X_2)$ признаков $X_1$ и $X_2$ в генеральной совокупности:
\begin{equation}
Cov(X_1,X_2) = E(X_1X_2) - E(X_1)E(X_2) = 0 - 0,16 * 0,23 = -0,0368
\end{equation}

Всего в городе $N$ избирателей (объем генеральной совокупности), тогда объем бесповторной выборки $n = 0,15N$.
Пусть $\overline X_1$ и $\overline X_2$ - выборочные средние сторонников партий А и В соответственно, тогда $n_B$ = $n\overline X_2$, а $\hat p_A = \overline X_1$.


Находим ковариацию $Cov(n_B,\hat p_A)$ для бесповторной выборки объема $n = 0,15N$:
\begin{equation}
Cov(n_B,\hat p_A) = Cov(\overline X_1,n\overline X_2) = n*Cov(\overline X_1,\overline X_2)= n*\frac{1}{n}Cov(X_1,X_2)\frac{N-n}{N-1} = Cov(X_1,X_2)\frac{N-n}{N-1}\approx Cov(X_1,X_2)\frac{N-n}{N} = Cov(X_1,X_2)(1 - \frac{n}{N}) = Cov(X_1,X_2)(1 - \frac{0,15N}{N}) = 0,85*Cov(X_1,X_2) = -0,85*0,0368 = -0,03128
\end{equation}
Ответ: $Cov(\overline n_B,\hat p_A)\approx -0,03128$.

### 3.9
Интервальное распределение имеет вид:
\begin{array}{rr} \hline
X_1 &13 &33 &53\\ \hline
P(A) &0,2 &0,6 &0,2\\ \hline
\end{array}

Последовательно находим сначала интервальное среднее $\overline x^*$:
\begin{equation}
\overline x^* = \frac{1}{n}\sum\limits_{i = 1}^kx_i^*n_i = 13*0,2 + 33*0,6 + 53*0,2 = 33
\end{equation}
Затем интервальную дисперсию $\sigma^{*2}$:
\begin{equation}
\sigma^{*2} = \frac{1}{n}\sum\limits_{i = 1}^k(x_i^* - \overline x^*)n_i = (13-33)^2*0,2 + (33-33)^2*0,6 + (53-33)^2*0,2 = 400*0,2*2 = 400*0,4 = 160
\end{equation}
В данном случае концы интервалов образуют арифметическую прогрессию с шагом $h = 20$, следовательно, применяя поправку Шеппарда для интервального распределения, получим:
\begin{equation}
\sigma^2 \approx\sigma^{*2}-\frac{1}{12}h^2 => \sigma^2 \approx 160 - \frac{400}{12} \approx 126,7
\end{equation}
Значит, искомая ошибка в бесповторной выборке объема $n = 25$ равна:
\begin{equation}
\sigma_{\overline X} = \sqrt{\frac{\sigma^2}{n}\frac{N-n}{N-1}} = \sqrt{\frac{126,7}{25}\frac{100-25}{100-1}} = \sqrt{\frac{126,7}{25}\frac{75}{99}} = \sqrt{\frac{126,7}{33}} \approx 1,96
\end{equation}
Ответ: $\sigma_{\overline X} = 1,96$.

### 3.10

### 3.11
Введем случайную величину $X$ - индикатор (признак) наличия/отсутствия прививки у случайно взятого жителя и случайную величину $Y$ - индикатор (признак) того, что случайно взятый житель не делал прививку и при этом не заболел.

Тогда распределение $X$ имеет вид:
\begin{array}{rr} \hline
X &0 &1\\ \hline
P(X) &0,38 &0,62\\ \hline
\end{array}
А распределение $Y$ имеет:
\begin{array}{rr} \hline
Y &0 &1\\ \hline
P(Y) &0,88 &0,12\\ \hline
\end{array}
Тогда вероятность, заключающегося в том, что не сделавший прививку житель остался здоров, есть $P(Y) = 0,12$.
Вычислим ковариацию $Cov(X,Y)$ признаков $X$ и $Y$ в генеральной совокупности:
\begin{equation}
Cov(X,Y) = E(XY) - E(X)E(Y) = E(XY) - 0,38*0,12 = E(XY) - 0,0456 = 0 - 0,0456 = -0,0456
\end{equation}
Всего в городе $N$ жителей (объем генеральной совокупности), тогда объем бесповторной выборки $n = 0,15N$.
Пусть $\overline X$ и $\overline {Y}$ - выборочные средние числа жителей, сделавших прививку, и числа жителей, не сделавших ее, но при том не заболевших, соответственно, тогда $n_B$ = $n*\overline {Y}$, а $\hat p_A = \overline X$.


Находим ковариацию $Cov(n_B,\hat p_A)$ для бесповторной выборки объема $n = 0,15N$:
\begin{equation}
Cov(n_B,\hat p_A) = Cov(\overline X,n\overline Y) = n*Cov(\overline X,\overline {Y})= n*\frac{1}{n}Cov(X,Y)\frac{N-n}{N-1} = Cov(X,Y)\frac{N-n}{N-1}\approx Cov(X,Y)\frac{N-n}{N} = Cov(X,Y)(1 - \frac{n}{N}) = Cov(X,X|Y)(1 - \frac{0,15N}{N}) = 0,85*Cov(X,Y) = -0,85*0,0456  = -0,03876
\end{equation}
Ответ: $Cov(\overline n_B,\hat p_A)\approx -0,03876$.

Распределение количества $x_i$ очков, выпавших на грани игральной симметричной кости после одного подбрасывания имеет следующий вид:
\begin{array}{rr} \hline
x_i &1 &2 &3 &4 &5 &6\\ \hline
P(x_i) &1/6 &1/6 &1/6 &1/6 &1/6 &1/6\\ \hline
\end{array}

Тогда $\overline x_i = E(x_i) = \frac{1}{6}*(1+2+3+4+5+6) = \frac{21}{6}$


Если кость подбрасывается n раз, то исход (выпадение определенного количества очков от 1 до 6) не зависит от предыдущих исходов, следовательно:


$\overline {nx_i} = E(n*x_i) = n*E(x_i) = \frac{21}{6}n$


При $n = 1000$:
$\overline {nx_i} = \frac{21}{6}*1000 = 3500$

Согласно классической центральной предельной теореме, сумма $n$ независимых одинаково распределённых случайных величин имеет распределение близкое к $N(n\mu,n\sigma_2)$ или, что эквивалентно (при большом $n$):
\begin{equation}
\overline X \sim N(\mu,\frac{\sigma_2}{n})
\end{equation}

```python
'''Необходимые статистические библиотеки'''
#%pylab inline
from pylab import *
import math
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy import special
from matplotlib import mlab
```

```python
N = 1000
n = 1000
a = []
for j in range(N):
    d = np.random.randint(1,7,n)
    a.append(np.mean(d))
    
k = 1 + math.floor(math.log2(len(a)))

mu = 3.5
va = 35/12
s = np.sqrt(va)/np.sqrt(n)

x = np.linspace(mu-3*s,mu+3*s,1000)
y = stats.norm.pdf(x,mu,s)

fig = plt.figure()
plt.plot(x,y)
plt.hist(a,bins=k,density=True)

```

```python
#%pylab inline
from pylab import *
import math
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy import special
from matplotlib import mlab

dice = [1,2,3,4,5,6]
sig = np.std(dice)
m = np.mean(dice)
N = 1000
n = 1000
a = []
for j in range(1,N):
    d = np.random.randint(1,7,n)
    a.append(np.mean(d))
    
aNew = []
for i in range(0,len(a)):
    aNew.append(math.sqrt(n)*(a[i]-3.5))
    
k =  1 + math.floor(math.log2(len(aNew)))

m1 = m - 3.5
print(m1)
plt.hist(aNew,bins=k,density=True)
x = np.arange(-3*sig,3*sig,0.01)
y = stats.norm.pdf(x, m1, sig)
plt.plot(x,y)
```
